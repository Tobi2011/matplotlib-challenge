```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
clinical_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
trial_data = pd.merge(mouse_data,clinical_data,
                     on = 'Mouse ID')

# Display the data table for preview
trial_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor_volume = trial_data.groupby(['Drug', 'Timepoint']).mean()
# Convert to DataFrame
mean_tumor_volume = pd.DataFrame(mean_tumor_volume)\
                    .reset_index()\
                    .drop(['Metastatic Sites'],axis = 1)
# Preview DataFrame
mean_tumor_volume.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sem_tumor_volumes = trial_data.groupby(['Drug', 'Timepoint']).sem()
# Convert to DataFrame
sem_tumor_volumes = pd.DataFrame(sem_tumor_volumes)\
                    .reset_index()\
                    .drop(['Metastatic Sites','Mouse ID'],axis = 1)
                
# Preview DataFrame
sem_tumor_volumes.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
sem_tumor_clean = sem_tumor_volumes.pivot('Timepoint','Drug','Tumor Volume (mm3)')
mean_tumor_clean = mean_tumor_volume.pivot('Timepoint','Drug','Tumor Volume (mm3)')
# Preview that Reformatting worked
mean_tumor_clean
```

```python
#get column names
header = list(mean_tumor_clean.columns.values)
header
mean_tumor_clean.iloc[0:len(mean_tumor_clean),0]
np.random.rand(3,)
```

```python
# Generate the Plot (with Error Bars)
fig, (drug) = plt.subplots(figsize = (10,10))

#plot
markers = ['o','s','D']
for x in range(len(header)):
    
    drug.errorbar(np.arange(0,50,5), mean_tumor_clean.iloc[:,x],
            yerr = sem_tumor_clean.iloc[:,x],
            fmt = random.choice(markers),
            markersize = 7,
            linestyle = '--',
            linewidth = .7,
            color = np.random.rand(3,),
            alpha = np.random.uniform(.9,1.0),
            label = "{}".format(header[x]))
    drug.legend(loc = 'upper left')

#plot labels and limits
drug.set_title('Tumor Response for Treatment')
drug.set_xlabel('Time (Days)')
drug.set_ylabel('Tumor Volume (mm3)')

drug.set_xlim([0,50])
drug.set_ylim([30,75])    

# Save the Figure
plt.savefig('Plots/Tumor Response for Treatment.png')
plt.grid()
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_metastatic = trial_data.groupby(['Drug', 'Timepoint']).mean()

# Convert to DataFrame
mean_metastatic = pd.DataFrame(mean_metastatic)\
                    .reset_index().drop(['Tumor Volume (mm3)'],axis=1)
                    
# Preview DataFrame
mean_metastatic.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
sem_metastatic = trial_data.groupby(['Drug', 'Timepoint']).sem()
# Convert to DataFrame
sem_metastatic = pd.DataFrame(sem_metastatic)\
                    .reset_index()\
                    .drop(['Tumor Volume (mm3)','Mouse ID'],axis = 1)
                    
# Preview DataFrame
sem_metastatic.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
sem_metastatic_clean = sem_metastatic.pivot('Timepoint','Drug','Metastatic Sites')
mean_metastatic_clean = mean_metastatic.pivot('Timepoint','Drug','Metastatic Sites')

# Preview that Reformatting worked
mean_metastatic_clean
```

```python
# Generate the Plot (with Error Bars)
fig, (drug) = plt.subplots(figsize = (10,10))

markers = ['o','s','D']

for x in range(len(header)):
    
    drug.errorbar(np.arange(0,50,5), mean_metastatic_clean.iloc[:,x],
            yerr = sem_metastatic_clean.iloc[:,x],
            fmt = random.choice(markers),
            markersize = 7,
            linestyle = '--',
            linewidth = .7,
            elinewidth = 1,
            color = np.random.rand(3,),
            alpha = np.random.uniform(.9,1.0),
            label = "{}".format(header[x]))
    drug.legend(loc = 'upper left')
    
drug.set_title('Metastatic Spread During Treatment')
drug.set_xlabel('Time (Days)')
drug.set_ylabel('Met. Sites')

drug.set_xlim([-.5,46])
drug.set_ylim([-.1,4])

# Save the Figure
plt.savefig('Plots/Metastatic Spread During Treatment.png')
plt.grid()
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival = trial_data.groupby(['Drug','Timepoint']).count()

# Convert to DataFrame
survival = pd.DataFrame(survival)[['Mouse ID']].reset_index()
survival = survival.rename(columns = {'Mouse ID': "Mouse Count"})

# Preview DataFrame
survival.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_clean = survival.pivot('Timepoint','Drug','Mouse Count')

# Preview that Reformatting worked
survival_clean
```

```python
# Generate the Plot (Accounting for percentages)
fig, (drug) = plt.subplots(figsize = (10,10))

drug.set_title('Survival During Treatment')
drug.set_xlabel('Time (Days)')
drug.set_ylabel('Survival Rate (%)')

markers = ['o','s','D']

#plot

for x in range(len(header)):
    
    drug.plot(np.arange(0,50,5), survival_clean.iloc[:,x].div(survival_clean.iloc[0,x]) *100,
            marker = random.choice(markers),
            markersize = 7,
            linestyle = '--',
            linewidth = .7,
            color = np.random.rand(3,),
            alpha = np.random.uniform(.9,1.0),
            label = "{}".format(header[x]))
    drug.legend(loc = 'lower left')

# Save the Figure
plt.savefig('Plots/Survival During Treatment.png')
plt.grid()
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
tumor_changes =(mean_tumor_clean.iloc[\
                len(mean_tumor_clean)-1,:] \
                - mean_tumor_clean.iloc[0,:] )\
                / mean_tumor_clean.iloc[0,:] \
                * 100

# Display the data to confirm
print(tumor_changes)
```

```python
# Store all Relevant Percent Changes into a Tuple
fig, summary = plt.subplots(figsize = (15,10))

# Splice the data between passing and failing drugs
passing = tumor_changes[tumor_changes < 0].reset_index()
failing = tumor_changes[tumor_changes >= 0].reset_index()

pass_plot = summary.bar('    ' + passing.iloc[:,0], passing.iloc[:,1],
           width = 1,
           align = 'edge',
           color = 'g')

fail_plot = summary.bar('    ' + failing.iloc[:,0], failing.iloc[:,1],
           width = 1,
           align = 'edge',
           color = 'r')

#tick label alignment
plt.setp( summary.xaxis.get_majorticklabels(), ha="left" )

#grid in plot
summary.grid()

# labels
summary.set_title('Tumor Change Over 45 Day Treatment')
summary.set_xlabel('Drug')
summary.set_ylabel('% of Tumor Volume Change')

# Use functions to label the percentages of changes
for rect in pass_plot + fail_plot:
    height = rect.get_height()
    summary.text(rect.get_x() + rect.get_width()/2., 2*height/abs(height),
            '%d%%' % int(height),
            ha='center',
            va='center',
            color = 'white')

#plot figure format
summary.set_xlim([-.3,len(header)+.3])
summary.set_ylim([-25,60])
plt.tight_layout()

# Save the Figure
plt.savefig('Plots/Tumor Change Over 45 Day Treatment.png')

# Show the Figure
fig.show()
```
